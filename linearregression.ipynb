```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("percapitaincome.csv")
df
```

Plotting scatterplot

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Year vs Per Capita Income")
plt.scatter(df['year'], df['per capita income (US$)'],color='red',marker='+')
```

Training the model

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

Predict per capita income

```python
reg.predict([[2004]])
```

```python
reg.coef_ # this is m in y=mx+b
```

```python
reg.intercept_ # this is b i.e intercept
```

 y=m*x + b Here m=828.46507522 ,b=-1632210.7578554575 , x is the year =2004

```python
828.46507522 *2004-1632210.7578554575
```
